```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#여러개의 subplot
def ReLU(x):
    return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = ReLU(x)

fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')

plt.title("ReLU")
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(0.0, color = 'gray', linestyle = "--")
plt.axhline(0.0, color = 'gray', linestyle = "--")
plt.plot(x,y)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)

fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')

plt.title("Sigmoid")
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(0.0, color = 'gray', linestyle = "--")
plt.axhline(0.0, color = 'gray', linestyle = "--")
plt.plot(x,y)
plt.show()
```

```python
def tanh(x):
    p_exp_x = np.exp(x)
    m_exp_x = np.exp(-x)
    
    y = (p_exp_x-m_exp_x)/(p_exp_x+m_exp_x)
    
    return y

x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)

fig = plt.figure(figsize=(10,7))
fig.set_facecolor('white')

plt.plot(x,y)
plt.title("tanh")
plt.xlabel('x')
plt.ylabel('y')

plt.yticks([-1.0, 0.0, 1.0])
plt.axvline(0.0, color='k')
ax=plt.gca()
ax.yaxis.grid(True)

plt.show()
```

```python
#시그모이드 함수 미분
def diff_sigmoid(x):
    return 1/(1+np.exp(-x)) * (1 - (1/(1+np.exp(-x))))

#tanh미분
def diff_tanh(x):
    return 4/(np.exp(2*x) + 2 + np.exp(-2*x))

x = np.arange(-10.0, 10.0, 0.1)
y1 =diff_sigmoid(x)
y2 = diff_tanh(x)

fig = plt.figure(figsize=(10,5))


plt.plot(x,y1, label = "diff_sigmoid")
plt.plot(x,y2, label = "diff_tanh")
plt.title("sigmoid vs math")
plt.xlabel('x')
plt.ylabel('y')

plt.ylim(-0.5,2)
plt.xlim(-7,7)

plt.legend(loc = "upper right")

plt.axvline(0.0, color='k')
ax=plt.gca()
ax.yaxis.grid(True)
ax.xaxis.grid(True)

plt.show()
```

```python
def Leaky_ReLU(x):
    
    return np.maximum(0.01*x, x)

x = np.arange(-100.0, 100.0, 0.1)
y = Leaky_ReLU(x)

fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')

plt.ylim(-5,20)
plt.title("Leaky ReLU")
plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0.0, color='gray')
plt.axhline(0.0, color = 'gray')
plt.plot(x,y)

plt.show()
```

```python
def Leaky_ReLU(x):
    
    return np.maximum(0.01*x, x)

def PReLU(x,a):
    
    return np.maximum(a*x, x)

x = np.arange(-100.0, 100.0, 0.1)
y1 = Leaky_ReLU(x)
y2 = PReLU(x,0.05)

fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')

plt.ylim(-5,20)
plt.title("Leaky ReLU &PReLU")
plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0.0, color='gray')
plt.axhline(0.0, color = 'gray')

plt.plot(x,y1, c="blue", label="Leaky ReLU")
plt.plot(x,y2, c="green", label="PReLU")
plt.legend(loc="upper right")
plt.show()
```

```python
def ELU(x, a=1):
    return (x>0)*x + (x<=0)*(a*(np.exp(x)-1))

x=np.arange(-30.0, 30.0, 0.1)
a=1
y = ELU(x,a)

plt.ylim(-3,25)
plt.xlim(-20,20)
plt.title("ELU")
plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0.0, color='gray', linestyle="--")
plt.axhline(0.0, color = 'gray', linestyle="--")

plt.plot(x,y)

plt.show()
```

```python
def softmax(x):
    exp_x = np.exp(x)
    result = exp_x / np.sum(exp_x)
    
    return result

x=np.arange(-5.0, 5.0, 0.1)
y = softmax(x)
print(x)
fig = plt.figure(figsize=(10,7))
fig.set_facecolor('white')

plt.plot(x,y)
plt.ylim(0,0.1)
plt.title("Softmax")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
